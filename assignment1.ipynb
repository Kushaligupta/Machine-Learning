```python
x = np.array([1,2,3,4,5,1,2,1,1,1])
most_freq_x = np.bincount(x).argmax()
indices_x = np.where(x == most_freq_x)[0]
print(most_freq_x,indices_x)

y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
most_freq_y = np.bincount(y).argmax()
indices_y = np.where(y == most_freq_y)[0]
print(most_freq_y,indices_y)
```

```python
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print(gfg.sum())
print(gfg.sum(axis=1))
print(gfg.sum(axis=0))
```

```python
n_array = np.array([[55, 25, 15], [30, 44, 2], [11, 45, 77]])

diag_sum = np.trace(n_array)
print(diag_sum)

eigen_values, eigen_vectors = np.linalg.eig(n_array)
print(eigen_values)
print(eigen_vectors)

inverse_matrix = np.linalg.inv(n_array)
print(inverse_matrix)

det = np.linalg.det(n_array)
print(det)
```

```python
p1 = np.array([[1, 2], [2, 3]])
q1 = np.array([[4, 5], [6, 7]])
product1 = np.dot(p1, q1)
cov1 = np.cov(p1, q1, rowvar=False)
print(product1)
print(cov1)

p2 = np.array([[1, 2], [2, 3], [4, 5]])
q2 = np.array([[4, 5, 1], [6, 7, 2]])
product2 = np.dot(p2, q2)
cov2 = np.cov(p2, rowvar=False)
print(product2)
print(cov2)
```

```python
x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])

inner_product = np.inner(x, y)
outer_product = np.outer(x, y)
cartesian_product = np.array(np.meshgrid(x.flatten(), y.flatten())).T.reshape(-1, 2)

print(inner_product)
print(outer_product)
print(cartesian_product)
```

```python
array = np.array([[1, -2, 3], [-4, 5, -6]])

print(np.abs(array))

flat = array.flatten()
print(np.percentile(flat, [25, 50, 75]))
print(np.percentile(array, [25, 50, 75], axis=0))
print(np.percentile(array, [25, 50, 75], axis=1))

print(flat.mean())
print(np.median(flat))
print(np.std(flat))

print(array.mean(axis=0))
print(np.median(array, axis=0))
print(np.std(array, axis=0))

print(array.mean(axis=1))
print(np.median(array, axis=1))
print(np.std(array, axis=1))
```

```python
from PIL import Image
import numpy as np

def img_to_array(path):
    img = Image.open(path)
    arr = np.array(img)

    if len(arr.shape) == 2:
        np.savetxt("grayscale_image.txt", arr, fmt='%d')
    else:
        for i, channel in enumerate(['R', 'G', 'B']):
            np.savetxt(f"{channel}_channel.txt", arr[:, :, i], fmt='%d')
    return arr
    
img_array = img_to_array("pexels-mccutcheon-1191710.jpg") 
import numpy as np
try:
    gray = np.loadtxt("grayscale_image.txt")
    print("Loaded grayscale image shape:", gray.shape)
except:
    print("Grayscale image not found.")

try:
    r = np.loadtxt("R_channel.txt")
    g = np.loadtxt("G_channel.txt")
    b = np.loadtxt("B_channel.txt")
    rgb = np.stack((r, g, b), axis=-1).astype(np.uint8)
    print("Loaded RGB image shape:", rgb.shape)
except:
    print("RGB channels not found.")

import matplotlib.pyplot as plt

if 'gray' in locals():
    plt.imshow(gray, cmap='gray')
    plt.title("Grayscale Image")
    plt.axis('off')
    plt.show()

if 'rgb' in locals():
    plt.imshow(rgb)
    plt.title("Reconstructed RGB Image")
    plt.axis('off')
    plt.show()
```
